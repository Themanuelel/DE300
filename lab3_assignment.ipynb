```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random

plt.rcParams.update(**{'figure.dpi':150})
plt.style.use('ggplot') # can skip this - plots are more visually appealing with this style
```

```python
use_cols = [
    "Year", "Month", "DayofMonth", "DayOfWeek", "DepTime", "CRSDepTime",
    "AirTime", "CRSArrTime", "Reporting_Airline","Flight_Number_Reporting_Airline", 
    "Tail_Number", "ActualElapsedTime", "CRSElapsedTime", "AirTime", "ArrDelay", "DepDelay", "Origin", "Dest",
    "Distance", "Flights", "TaxiIn", "TaxiOut", "Cancelled",
    "CancellationCode", "Diverted", "CarrierDelay", "WeatherDelay", "NASDelay", "SecurityDelay",
    "LateAircraftDelay"
]
```

```python
#jan = pd.read_csv(
#    'On_Time_Reporting_Carrier_On_Time_Performance_2019_1.csv',usecols=use_cols
#)
#print(f'Number of observations in Jan 2019: {jan.shape[0]}')
#jan.head(5)

total_rows = sum(1 for _ in open('On_Time_Reporting_Carrier_On_Time_Performance_2019_1.csv')) - 1
sample_size = 100000
skip_indices = sorted(random.sample(range(1, total_rows + 1), total_rows - sample_size))

reader = pd.read_csv(
    'On_Time_Reporting_Carrier_On_Time_Performance_2019_1.csv',
    usecols=use_cols,
    skiprows=skip_indices,
    chunksize=1000
)
jan = pd.concat([chunk for chunk in reader])
# print(jan.shape[0])
```

```python
jan.head(5)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 4))

#AirTime histogram
sns.histplot(jan['AirTime'], ax=axs[0], bins=20)
axs[0].set_title('AirTime')
axs[0].set_xlabel('AirTime (minutes)')
axs[0].set_ylabel('Frequency')

#Distance histogram
sns.histplot(jan['Distance'], ax=axs[1], bins=20)
axs[1].set_title('Distance')
axs[1].set_xlabel('Distance (miles)')
axs[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=jan, x='AirTime', y='Distance')
plt.title('AirTime vs Distance')
plt.xlabel('AirTime')
plt.ylabel('Distance')
plt.show()
```

## The above plots show the distribution of airtime and distances for flights and compares them together. Both distributions for airtime and distance are right skewed. As can be seen in the AirTime vs Distance plot, there is a clear positive correlation between the distance of a flight and the length of time it takes to travel. This makes sense because it should take longer to travel a farther distance. 

```python
flight_per_day_of_week = jan['DayOfWeek'].value_counts().sort_index()

plt.figure(figsize=(8, 6))
flight_per_day_of_week.plot(kind='bar', color='skyblue')
plt.title('Number of Flights per Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Number of Flights')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
cancelled_flights = jan[jan['Cancelled'] == 1]
cancelled_flights_per_day_of_week = cancelled_flights['DayOfWeek'].value_counts().sort_index()

plt.figure(figsize=(8, 6))
cancelled_flights_per_day_of_week.plot(kind='bar', color='salmon')
plt.title('Number of Cancelled Flights per Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Number of Cancelled Flights')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
total_flights_per_day_of_week = jan['DayOfWeek'].value_counts().sort_index()
cancelled_percentage_per_day_of_week = (cancelled_flights_per_day_of_week / total_flights_per_day_of_week) * 100

plt.figure(figsize=(8, 6))
cancelled_percentage_per_day_of_week.plot(kind='bar', color='salmon')
plt.title('Percentage of Cancelled Flights per Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Percentage of Cancelled Flights')
plt.xticks(rotation=0)
plt.ylim(0, 6)  # Set y-axis limit to 6 percent
plt.tight_layout()
plt.show()
```

## The above plots show the number of flights held per day of the week, the amount of cancelled flights per day of the week, and the percentage of flights that are cancelled per each day of the week. The first two plots show that most flights occur on the fourth day of the week (Thursday) while most cancelled flights occur on the third day of the week (Wednesday). The third plot shows that around 3% of all flights are cancelled on any given day, but the fifth day of the week (Friday) has a substantially smaller percentage of cancelled flights. This is surprising, since many sites online actually say that Friday has the most cancelled flights.
